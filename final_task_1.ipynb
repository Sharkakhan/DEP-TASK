```python
from sklearn.preprocessing import StandardScaler
selected_columns = ['TotalAmount', 'ProductCategory', 'PaymentMethod']
data_selected = df[selected_columns]
```

```python
data_encoded = pd.get_dummies(data_selected, columns=['ProductCategory', 'PaymentMethod'])
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_encoded)
```

```python
data_scaled_df = pd.DataFrame(data_scaled, columns=data_encoded.columns)
data_scaled_df.head()
```

```python
inertia = []
for n in range(1, 11):
  kmeans = KMeans(n_clusters=n, random_state=0)
  kmeans.fit(data_scaled_df)
  inertia.append(kmeans.inertia_)

```

```python
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('inertia')
plt.title('Elbow Method')
plt.show()
```

```python
optimal_clusters = 5
kmeans = KMeans(n_clusters=optimal_clusters, random_state=0)
kmeans.fit(data_scaled_df)

df['Cluster'] = kmeans.labels_
print(df.head())
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df['TotalAmount'], df['Cluster'], c=df['Cluster'], cmap='rainbow')
plt.xlabel('TotalAmount')
plt.ylabel('Cluster')
plt.title('TotalAmount vs Cluster')
plt.colorbar()
plt.show()
```
